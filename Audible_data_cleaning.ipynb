#### Data Cleaning:

1. Remove Written By from Author column:

```python
df['author']
```

```python
df['author'] = df['author'].apply(lambda x: x.split(':')[-1])
df.head()
```

2. Remove Narratedby from narrator Column:

```python
df['narrator'] = df['narrator'].apply(lambda x: x.split(':')[-1])
df.head()
```

3. Convert time to minutes:

```python
df['time'].unique()
```

```python
def convert_to_minutes(time_str):
    parts = time_str.split()
    hours = 0
    minutes = 0
    
    for i in range(len(parts)):
        if parts[i] == 'hrs' or parts[i] == 'hr':
            hours = int(parts[i-1])
        elif parts[i] == 'mins' or parts[i] == 'min':
            minutes = int(parts[i-1])
    
    total_minutes = hours * 60 + minutes
    return total_minutes
```

```python
df['time'] = df['time'].apply(lambda x: convert_to_minutes(x))
df.head()
```

4. Convert releasedate dtype:

```python
df['releasedate'] =  pd.to_datetime(df['releasedate'])
df.head()
```

5. Seprate stars and total_ratings:

```python
df.tail()
```

```python
def ratings(value):
    if 'ratings' in value:
        stars_index = value.index("stars")
        # Extract the substring after "stars" and remove any leading/trailing whitespace
        rating_text = value[stars_index + len("stars"):].strip()
        # Extract the numeric rating from the rating text
        rating = int(''.join(filter(str.isdigit, rating_text)))
        return rating
    else:
        return 0
```

```python
df['ratings'] = df['stars'].apply(lambda x: ratings(x))
df.head()
```

```python
def Stars(s):
    values = s.split(' ')
    if values[0] == 'Not' or values[0] == 'not':
        return 0
    else:
        return float(values[0])
```

```python
df['stars'] = df['stars'].apply(lambda x : Stars(x))
df.head()
```

6. Convert dtype of price:

```python
def convert_price(p):
    price = p.replace(',','')
    if price == 'Free':
        return 0
    else:
        return float(price)
```

```python
df['price'] = df['price'].apply(lambda x: convert_price(x))
df.head()
```

```python
df.to_csv('audible_cleaned.csv',index=0)
```
